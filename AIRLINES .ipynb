```python
data['airline'].value_counts(ascending = True).plot.barh(color = ['lightgreen' , 'lightblue'])

plt.title('Airlines with frequences')

plt.xlabel('number of flights')

plt.ylabel('Airline')


plt.show()

```

```python
data['arrival_time'].value_counts()
```

```python
data['departure_time'].value_counts()
```

```python
plt.figure(figsize = (14,4))
plt.subplot(1,2,1)
plt.bar( data['departure_time'].value_counts().index , data['departure_time'].value_counts().values , color = ['g' ,'b'] )
plt.title("Departure_Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")
plt.subplot(1,2,2)
plt.bar( data['arrival_time'].value_counts().index , data['arrival_time'].value_counts().values , color = ['y' , 'r'])
plt.title("Arrival_Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")
plt.show()
```
